
## Ejercicio 1

Crea una nueva columna en tu dataset.

Ideas:

- combinar columnas
- crear una métrica nueva
- convertir unidades

Motivación:

> ¿Qué nueva variable podría ayudarte a entender mejor los datos?


````python
# Tu solución aquí

# ( 1 )

# df['popularity'] = df['id'].astype(str) + ' ' + df['original_title']

df["salary_avg"] = df[["price_start", "price_end"]].mean(axis=1)

df.head()
````


## Ejercicio 2

Filtra el dataset usando una condición.


## Ejercicio 3 — Múltiples condiciones

Podemos combinar condiciones usando:

- `&` (AND)
- `|` (OR)


````python
# Ejemplo
# df[(df["col1"] > 10) & (df["col2"] < 50)]
````

````python
# Tu solución aquí
df[(df["city"].str.lower() == "madrid") & (df["salary_avg"] > 30000)]
````


# Agrupar datos

La función **groupby()** permite resumir información por categorías.

Ejemplo:

```python

```


````python
# Ejemplo genérico
# df.groupby("columna_categoria").mean()
````


## Ejercicio 4

Agrupa el dataset por una columna categórica y calcula una estadística.

Motivación:

> ¿Cómo cambian los valores dependiendo de la categoría?


````python
# Tu agrupación aquí
df.groupby("province")["salary_avg"].mean().sort_values(ascending=False)
````


## Ejercicio 5 — Agrupación por varias columnas


````python
# Tu solución aquí
df.groupby(["province", "type"])["salary_avg"].mean().reset_index().sort_values(by="salary_avg", ascending=False)
````


# Desafío final

Combina varias operaciones:

1. Filtra datos
2. Agrupa
3. Ordena el resultado


````python
# Tu análisis final aquí

df_filtered = df[(df["salary_avg"].notna()) & (df["type"] == "Full-time")]

resultado_final = df_filtered.groupby("province")["salary_avg"].mean().reset_index()

resultado_final.sort_values(by="salary_avg", ascending=False)
````

# 1. Introducción a Pandas

## Teoría

**Pandas** es una de las librerías más utilizadas en Python para:

- Análisis de datos
- Limpieza de datasets
- Transformación de datos
- Exploración inicial

Las dos estructuras principales de Pandas son:

### Series  → estructura de datos unidimensional. 
Para crear una serie El método más común es pasar una lista: 
```
s = pd.Series([1, 3, 5, 7])
```

### DataFrame  → tabla bidimensional (similar a Excel o SQL)
Para crear un dataframe es necesario pasar un diccionario:
```
df = pd.DataFrame({'Col1': [1, 2], 'Col2': ['a', 'b']})
```

Pandas permite trabajar con datos de forma muy eficiente.

# 2. DataFrame

## Teoría

Un **DataFrame** es la estructura principal de Pandas.

Se puede imaginar como una **tabla**:

- Filas → observaciones
- Columnas → variables

Cada columna puede tener un tipo de dato distinto.

Los DataFrames permiten:

- Filtrar datos
- Agrupar datos
- Calcular estadísticas
- Transformar columnas

````python
data = {
    "Nombre": ["Ana", "Luis", "Marta"],
    "Edad": [25, 30, 22],
    "Ciudad": ["Madrid", "Barcelona", "Valencia"]
}

df_ejemplo = pd.DataFrame(data)
df_ejemplo
````

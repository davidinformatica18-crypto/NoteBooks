# 2. Gráfico de barras (Matplotlib)

Los **gráficos de barras** se utilizan para comparar cantidades entre categorías.

Ejemplo: número de pasajeros por clase.


```python
class_counts = df["pclass"].value_counts()
class_counts
```

```python
plt.bar(class_counts.index, class_counts.values)

plt.title("Número de pasajeros por clase")
plt.xlabel("Clase")
plt.ylabel("Número de pasajeros")

plt.show()
```

# 3. Gráfico de líneas

Los **gráficos de líneas** se utilizan para ver tendencias.


```python
age_counts = df["age"].value_counts().sort_index()

plt.plot(age_counts.index, age_counts.values)

plt.title("Distribución de edades")
plt.xlabel("Edad")
plt.ylabel("Número de pasajeros")

plt.show()
```

# 7. Boxplot

El **boxplot** permite visualizar:

- mediana
- cuartiles
- valores atípicos


```python
sns.boxplot(data=df, x="pclass", y="age")

plt.title("Edad por clase")

plt.show()
```

# 8. Heatmap de correlaciones

Un **heatmap** permite visualizar la correlación entre variables numéricas.


```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Matriz de correlación")

plt.show()
```
